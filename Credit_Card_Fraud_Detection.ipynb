Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#Loading the dataset to a Pandas DataFrame
credit_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML Projects/Credit Card Fraud Detection/creditcard.csv')
```

```python
# Checking the number of missing values in each columns
credit_data.isnull().sum()
```

```python
# Distribution of legit & fraudulant transactions
credit_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 --> Normal Transaction

1 --> Fraudulant Transaction

```python
# Separating the data for analysis
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the dataset
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing smaller distribution of Normal Transactions and Fraudulant Transactions

Number of Fraudulant Transactions = 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two datasets

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the dataset into features & target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splittingthe data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = metrics.accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = metrics.accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```
